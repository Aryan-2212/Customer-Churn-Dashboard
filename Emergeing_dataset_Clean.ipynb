```python
import pandas as pd

path = '/content/BankChurners.csv'

df = pd.read_csv(path)
df.columns = df.columns.str.strip()

print("Here's the head:")
df.head()
```

```python
df.columns
```

```python
df.isnull().sum()
df.isnull().mean() * 100
```

```python
print("Duplicate rows:", df.duplicated().sum())

df.drop_duplicates(inplace=True)
```

```python
df['Credit_Limit'] = df['Credit_Limit'].astype(float)
df['Avg_Open_To_Buy'] = df['Avg_Open_To_Buy'].astype(float)
```

```python
df.drop(columns=['CLIENTNUM'], errors='ignore', inplace=True)
```

```python
# Handle categorical missing values explicitly
cat_missing_cols = ['Education_Level', 'Marital_Status', 'Income_Category']

for col in cat_missing_cols:
    df[col] = df[col].replace('Unknown', 'Missing')
```

```python
# Create missing indicator flags
for col in cat_missing_cols:
    df[col + '_missing_flag'] = (df[col] == 'Missing').astype(int)
```

```python
# Convert target to binary
df['Attrition_Flag'] = df['Attrition_Flag'].map({
    'Existing Customer': 0,
    'Attrited Customer': 1
})
```

```python

print("\nFinal Missing Values:\n", df.isnull().sum())
print("\nFinal Dataset Shape:", df.shape)
print("\nData Types:\n", df.dtypes)
```

```python
X = df.drop('Attrition_Flag', axis=1)
y = df['Attrition_Flag']

cat_cols = X.select_dtypes(include='object').columns
num_cols = X.select_dtypes(include=['int64', 'float64']).columns
```

```python
cleaned_path = '/content/BankChurners_Cleaned.csv'
df.to_csv(cleaned_path, index=False)

print("Cleaned dataset saved successfully.")
```
